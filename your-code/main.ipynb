## Challenge 3: Hidden Cold Joke

Using Python, call Github API to find out the cold joke contained in the 24 secret files in the following repo:

https://github.com/ironhack-datalabs/scavenger

The filenames of the secret files contain .scavengerhunt and they are scattered in different directories of this repo. The secret files are named from .0001.scavengerhunt to .0024.scavengerhunt. They are scattered randomly throughout this repo. You need to search for these files by calling the Github API, not searching the local files on your computer.

#### 1. Find the secret files.

`````python
# your code here
params3 = {'recursive' : 1}
path_and_endpoint3 = 'repos/ironhack-datalabs/scavenger/git/trees/9308ccc8a4c34c5e3a991ee815222a9691c32476'
response3 = requests.get(f'{base_url}{path_and_endpoint3}', auth = (user, token), params = params3)
result3 = response3.json()
`````

`````python
print(json.dumps(result3, indent=4))
`````

`````python
## Get paths list
list_of_scavengers = []
for dict in result3['tree']:
    if '.scavengerhunt' in dict['path']:
        list_of_scavengers.append(dict['path'])

list_of_scavengers
`````

#### 2.  Sort the filenames ascendingly.

`````python
## Get files list
list_of_files = []
for i in list_of_scavengers:
    list_of_files.append(i[-18:]) 
`````

`````python
# Order them
ordered_list_of_files = sorted(list_of_files)
ordered_list_of_files
`````

#### 3. Read the content of each secret files into an array of strings.
Since the response is encoded, you will need to send the following information in the header of your request:
````python
headers = {'Accept': 'application/vnd.github.v3.raw'}
````

`````python
# your code here

## Generate a dictionary that contains the filename as key and the URL of the filename as value
URL_dict = {}
for dict in result3['tree']:
    if '.scavengerhunt' in dict['path']:
        URL_dict[dict['path'][-18:]] = dict['url']

## Retrieve the content of each file following the ascending order previously created
file_content = []
for i in ordered_list_of_files:
    url = URL_dict[i]
    headers = {'Accept': 'application/vnd.github.v3.raw'}
    response4 = requests.get(url, auth = (user, token), headers = headers)
    result4 = response4.content
    file_content.append(result4)

`````

#### 4. Concatenate the strings in the array separating each two with a whitespace.

`````python
# your code here

##Clean strings
new_content = []
for i in file_content:
    j = i.strip()
    k = j.decode('UTF-8')
    new_content.append(k)

##Concatenate strings
sep = ' '
joke = sep.join(new_content)
`````
